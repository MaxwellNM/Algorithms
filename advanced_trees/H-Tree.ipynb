```python
class Point ():
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def printP (self):
    print("Point ({} {}) ".format(str(self.x),str(self.y)))
    
#P1 = Point(2,3)
#P1.printP()
max_depth = 10
k= 2#math.sqrt(2)

def drawlines2(centerLef, centerRight, topLeft, botomLeft, topRight, botomRight, curdepth):
  print("--- depth-- ",curdepth)
  print("segXaxis ({},{}) ({},{}):".format(str(centerLef.x),str(centerLef.y),str(centerRight.x),str(centerRight.y)))
  print("segYLeftAxis ({},{}) ({},{}):".format(str(botomLeft.x),str(botomLeft.y),str(topLeft.x),str(topLeft.y)))
  print("segYRightAxis ({},{}) ({},{}):".format(str(botomRight.x),str(botomRight.y),str(topRight.x),str(topRight.y)))
  print("-------------------------------------- ")

def drawlines(centerLef, centerRight, topLeft, botomLeft, topRight, botomRight, curdepth):
  plt.grid()
  plt.plot([centerLef.x,centerRight.x],[centerLef.y,centerRight.y])
  plt.plot([botomLeft.x,topLeft.x],[botomLeft.y,topLeft.y])
  plt.plot([botomRight.x, topRight.x],[botomRight.y, topRight.y])

def drawHtree(center, depth, lngt ):
  assert (depth  > 0 and depth < max_depth ) and lngt > 0 ,"0 < depth < {} and lngt > 0 ".format(str(max_depth+1))
  
    
  if isinstance(center, Point):
      cur_lngt = lngt/k
      centerLef = Point(center.x - cur_lngt, center.y)
      centerRight = Point(center.x + cur_lngt, center.y)
      topLeft = Point(centerLef.x,centerLef.y + cur_lngt)
      topRight = Point(centerRight.x,centerRight.y + cur_lngt)
      botomLeft = Point(centerLef.x,centerLef.y - cur_lngt)
      botomRight = Point(centerRight.x,centerRight.y - cur_lngt)
      # print H-tree depth 1
      cur_depth = 1
      drawlines(centerLef, centerRight, topLeft, botomLeft, topRight, botomRight, cur_depth)
      drawRecursive(topLeft, botomLeft, topRight, botomRight, cur_depth+1, depth,  cur_lngt)

def drawRecursive(topLeft, botomLeft, topRight, botomRight, cur_depth, depth, cur_lngt):
  
    if cur_depth > depth :
      return

    #cur_depth = cur_depth + 1
    cur_lngt = cur_lngt /k

    count = (cur_depth)*4
    mid = topLeft
    for num in range(1,count+1):
      mid = Point(-5,-5)
      #check direction
      if num < count/4 :
        mid = topLeft
      elif num < count/2 :
        mid = botomLeft
      elif num < (count*3)/4 :
        mid = topRight
      else:
         mid = botomRight
      # calculate the curent coordinates from the mid
      CurcenterLef = Point(mid.x - cur_lngt, mid.y)
      CurcenterRight = Point(mid.x + cur_lngt, mid.y)
      CurtopLeft = Point(CurcenterLef.x,CurcenterLef.y + cur_lngt)
      CurtopRight = Point(CurcenterRight.x,CurcenterRight.y + cur_lngt)
      CurbotomLeft = Point(CurcenterLef.x,CurcenterLef.y - cur_lngt)
      CurbotomRight = Point(CurcenterRight.x,CurcenterRight.y - cur_lngt)
      drawlines(CurcenterLef, CurcenterRight, CurtopLeft, CurbotomLeft, CurtopRight, CurbotomRight, cur_depth)
      #Recursive call H-Tree
      drawRecursive(CurtopLeft, CurbotomLeft, CurtopRight, CurbotomRight, cur_depth+1, depth,  cur_lngt)



        
     

```

```python
plt.figure(figsize=(5,5))
start = time.time()
drawHtree(Point(0,0), 1, 4 )
print("Time = ",time.time()-start)
```
